```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("Raw_Reviews_WM.xlsx")
df.head()
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df["rating"]= df["r_rating"].apply(lambda x : int(x[0]))
df.drop("r_rating", axis=1, inplace=True)
df.head()
```

```python
df_title = df[["r_title","rating"]]
df_title.columns=["Review", "rating"]
df_title.head()
```

```python
df_rev = df[["r_review ","rating"]]
df_rev.columns=["Review", "rating"]
df_rev.head()
```

```python
df1= pd.concat([df_title, df_rev], axis=0).reset_index(drop=True)
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.dropna(inplace=True)
```

```python
df1.reset_index(drop=True).head()
```

```python
import warnings
warnings.filterwarnings(action="ignore")
```

```python
# Lets Consider rating of 1 ,2 and 3 as Negative review , and considering Negative review as 0
# Lets Consider rating of 4, 5 as Positive Review , and conidering Positive review as 1

df1["sentiment"]= df1["rating"].map({1: 0, 2:0, 3:0, 4: 1, 5:1})
df1.head()
```

```python
df1.sentiment.value_counts()
```

```python
import re
from nltk.stem import PorterStemmer , WordNetLemmatizer
from nltk import word_tokenize
from nltk.corpus import stopwords
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
def text_cleaner (text):
    cleaned= re.sub('[^a-zA-Z]', " ", text) 
    cleaned= cleaned.lower()
    cleaned = cleaned.split()
    cleaned= [lemmatizer.lemmatize (word) for word in cleaned]
    cleaned= ' '.join(cleaned)
    return cleaned
```

```python
df_final= df1[["Review", "sentiment"]]
df_final
```

```python
df_final["Cleaned"] = df_final["Review"].apply(lambda x : text_cleaner(x))
df_final
```

```python
df_final.drop("Review", axis=1, inplace=True)
df_final
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence  import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.preprocessing.text import one_hot 
from tensorflow.keras.layers import Dropout 
```

```python
voc_size= 10000
```

```python
X= df_final["Cleaned"]
```

```python
y= df_final["sentiment"]
```

```python
import tensorflow as tf
tf.__version__
```

```python
x = list(X)
x
```

```python
onehot = [one_hot(words, voc_size) for words in x]
onehot
```

```python
sent_length = 20
embedded = pad_sequences(onehot, maxlen=sent_length, padding='pre' )
embedded
```

```python
emmbedding_vectors= 40
model = Sequential()
model.add(Embedding(voc_size,20, input_length= sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1, activation= "sigmoid"))
model.compile(loss= "binary_crossentropy", optimizer = "adam", metrics= ["accuracy"])
print(model.summary())
```

```python
x_final =np.array(embedded)
y_final = np.array(y)
x_final.shape, y_final.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( x_final, y_final, test_size=0.33, random_state=42)
```

```python
model.fit(X_train,y_train, validation_data=(X_test,y_test), epochs=5, batch_size = 40)
```

```python
y_pred = (model.predict(X_test) > 0.5).astype("int32")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
